```python
# Simple Interest Calculator

print("This is a Simple interest calculator. All inputs are digits(e.g 1000, 4343.88)")
Principal = float(input("What is your Principal: "))
Time = float(input("What is your Time: "))
Rate = float(input("What is the Rate : "))
Simple_Interest = (Principal * Time * Rate)/100
amount = Simple_Interest + Principal
print(f"Simple Interest is {Simple_Interest} and the total amount is {amount}")
```
