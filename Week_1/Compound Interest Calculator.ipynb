```python
# Compound Interest Calculator

print("This is a Compound interest calculator. All inputs are digits(e.g 1000, 4343.88)")
Principal = float(input("What is your Principal: "))
Time = float(input("What is your Time: "))
Rate = float(input("What is the Rate : "))
Number = float(input("What is the number of times interest is compounded per year : "))
Amount = Principal*(1+(Rate/Number)**(Number*Time))
Compound_Interest = Amount - Principal
print(f"Compound Interest is {Compound_Interest} and the total amount is {Amount}")

```
